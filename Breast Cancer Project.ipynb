## Performing Dimensionaliy Reduction to Reduce the features
### >Rather than dropping the columns from our DataSet we reduce the dimensionality. In this, we try to combine these features in some way to maximize the amount of information. This is exactly he goal of Principal Component Analysis (PCA).

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=10)
```

```python
pca_data = pca.fit_transform(numerical_data)
```

```python
pca_data = pd.DataFrame(pca_data)
```

```python
pca_data
```

```python
pca.explained_variance_ratio_
```

```python
sns.set()
plt.plot(pca.explained_variance_ratio_)
```

```python
numerical_data = pca_data
```

```python
from sklearn.model_selection import train_test_split
```

```python
X, x_test, Y, y_test = train_test_split(numerical_data,output_data,test_size=0.3, random_state=5)
```

### Through Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X,Y)
```

```python
predictions = model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
score = accuracy_score(predictions,y_test)
```

```python
score
```

### Through Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
decision_model = DecisionTreeClassifier()
```

```python
decision_model.fit(X,Y)
```

```python
decision_predictions = decision_model.predict(x_test)
```

```python
decision_score = accuracy_score(decision_predictions,y_test)
```

```python
decision_score
```

### Through KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model = KNeighborsClassifier()
```

```python
knn_model.fit(X,Y)
```

```python
knn_predictions = knn_model.predict(x_test)
```

```python
knn_score = accuracy_score(knn_predictions,y_test)
```

```python
knn_score
```

### through Voting Classifier

```python
from  sklearn.ensemble import VotingClassifier
```

```python
dt = DecisionTreeClassifier()
knn = KNeighborsClassifier()
lg = LogisticRegression()
```

```python
voting_model = VotingClassifier(estimators=[
    ('dt', dt),
    ('knn', knn)], voting='hard')
```

```python
voting_model.fit(X, Y)
```

```python
voting_predictions = voting_model.predict(x_test)
```

```python
voting_score = accuracy_score(voting_predictions, y_test)
```

```python
voting_score
```

### Comparison of Logistic Score, Decision Tree Score, KNN Score AND Voting Classifier Score
### - through Graph Visualization

```python
prediction_dict = {
    'LogReg Score' : 100*score,
    'DT Score' : 100*decision_score,
    'KnnScore' : 100*knn_score,
    'VotingScore' : 100*voting_score
}
```

```python
prediction_dict
```
